## Face Mask Detection Without Transfer Learning

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, GlobalAveragePooling2D, Dense, BatchNormalization, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
```

```python
train_ds = tf.keras.utils.image_dataset_from_directory("face_mask_dataset/train", image_size=(128,128), batch_size=32)
val_ds   = tf.keras.utils.image_dataset_from_directory("face_mask_dataset/val", image_size=(128,128), batch_size=32)
test_ds  = tf.keras.utils.image_dataset_from_directory("face_mask_dataset/test", image_size=(128,128), batch_size=32)
```

```python
class_names = train_ds.class_names
print("Class names:", class_names)
```

```python
data_preprocess = tf.keras.Sequential(
    name="data_preprocess",
    layers=[layers.Rescaling(1.0/255)]
)
```

```python
train_ds = train_ds.map(lambda x, y: (data_preprocess(x), y))
val_ds = val_ds.map(lambda x, y: (data_preprocess(x), y))
test_ds = test_ds.map(lambda x, y: (data_preprocess(x), y))
```

```python
model = Sequential([
    
    Conv2D(32, kernel_size=(3,3), padding='same', activation='relu', input_shape=(128, 128,3)),
    Conv2D(32, kernel_size=(3,3), padding='same', activation='relu'),
    MaxPooling2D(),

    Conv2D(64, kernel_size=(3,3), padding='same', activation='relu'),
    Conv2D(64, kernel_size=(3,3), padding='same', activation='relu'),
    BatchNormalization(),
    MaxPooling2D(),

    Conv2D(128, kernel_size=(3,3), activation='relu'),
    Conv2D(128, kernel_size=(3,3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D(),

    Conv2D(256, kernel_size=(3,3), padding='same', activation='relu'),
    Conv2D(256, kernel_size=(3,3), padding='same', activation='relu'),
    BatchNormalization(),
    
    GlobalAveragePooling2D(),

    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.4),

    Dense(128, activation='relu'),
    BatchNormalization(),

    Dense(1, activation='sigmoid')

])
```

```python
model.summary()
```

```python
model.compile(
            optimizer = Adam(learning_rate=0.001),
            loss = 'binary_crossentropy',
            metrics = ['accuracy']
           )
```

```python
earlystopping = EarlyStopping(
    monitor='val_accuracy',       
    patience=5,               
    restore_best_weights=True
)

reduce_lr = ReduceLROnPlateau(factor=0.2, patience=3, verbose=1)
```

```python
Model = model.fit(train_ds, validation_data=val_ds, epochs=20, 
                  batch_size=64, verbose=1, callbacks=[earlystopping, reduce_lr]
                 )
```

```python
plt.plot(Model.history['accuracy'], label='Training Accuracy')
plt.plot(Model.history['val_accuracy'], label='Validation Accuracy')
plt.title('Face Mask Detection Accuracy - Training vs. Validation')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.show()
```

```python
pred = model.predict(test_ds)
```

```python
model.evaluate(test_ds)
```

```python
def load_and_preprocess_image(path):
    img = tf.keras.preprocessing.image.load_img(path, target_size=(128, 128))
    img = tf.keras.preprocessing.image.img_to_array(img)
    img = tf.expand_dims(img, axis=0)
    img = img / 255.0   # because you used Rescaling(1/255)
    return img
```

```python
img_path = "with_mask.jpeg"
input_image = load_and_preprocess_image(img_path)
prediction = model.predict(input_image)
print("Prediction:", prediction)

labels = (prediction > 0.5).astype(int)
if labels[0] == [0]:
    print('Image is with mask')
else:
    print('Image is without mask')
```

```python
img_path = "without_mask.jpeg"
input_image = load_and_preprocess_image(img_path)
prediction = model.predict(input_image)
print("Prediction:", prediction)

labels = (prediction > 0.5).astype(int)
if labels[0] == [0]:
    print('Image is with mask')
else:
    print('Image is without mask')
```
